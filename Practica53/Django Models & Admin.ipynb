# Código de creación de elementos granel o bulk


```python
from ecommerce.models import ProductModel
objects = []
products = []
for i in range(1,501):
	data = {
		"title":"Producto ejemplo {}".format(i),
		"price": i*100 +.99,
		"description":"Lorem ipsum",
		"color":"Blanco",
		"seller":"Marca Falsa 123"
	}
	products.append(data)

for product in products:
	objects.append(ProductModel(**product))
ProductModel.objects.bulk_create(objects, ignore_conflicts=True)
```
