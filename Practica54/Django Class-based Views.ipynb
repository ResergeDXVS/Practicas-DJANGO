```python
# Nota: El módulo lo guardé por error como projects

#src\projects\views.py
#Creación de lista protegida
class ProtectedListView(LoginRequiredMixin,TemplateTitleMixin,ListView):
    model = Product
    template_name = "/projects/product_list.html"
    title = "Mis productos"
    
    def get_queryset(self):
        return Product.objects.filter(user=self.request.user)
```
